#### 2.C [BONUS] Write a function to count the number of rows


```
def row_count(column_name):
    number_of_rows = 0
    return number_of_rows
```



````python
def row_count(df, column_name):
    """
    Counts the number of rows in a specified column of a DataFrame.
    
    Args:
        df (pandas.DataFrame): The DataFrame to count rows from.
        column_name (str): The name of the column to count rows from.
    
    Returns:
        int: The count of rows in the specified column.
    """
    return df[column_name].count()
````

````python
row_count(df, 'age')
````
